```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aus_plotting = pd.read_excel('AustraliaFiltered.xlsx')
```

```python
# Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants

data = aus_plotting.sort_values(by='Total', ascending=False).head()
plt.bar(data.Country.values, data.Total.values)
plt.xticks(rotation=90)
plt.show()

```

```python
# Find total immigrants from each category by area name and plot barchart for each area

df = aus_plotting.groupby(by=['AreaName'])

df1 = df.Total.sum()
print(df1)

print()
df1.plot(kind='bar')
plt.show()
```

```python
# Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot

world = aus_plotting.loc[:,'1980':'2008'].sum()
print(world)

world.plot(kind='line')
```

```python
# Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).

plt.figure(figsize=(15,6))
trend = aus_plotting.groupby(by = ['AreaName'])

Asia1 = trend.get_group('Asia').loc[:, '1980':'2008'].sum()
plt.plot(Asia1, label = 'Asia')

Africa1 = trend.get_group('Africa').loc[:, '1980':'2008'].sum()
plt.plot(Africa1, label = 'Africa')

Europe1 = trend.get_group('Europe').loc[:, '1980':'2008'].sum()
plt.plot(Europe1, label = 'Europe')

plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot immigration trend for India using line plot

india_trend = aus_plotting.index[aus_plotting.Country == 'India']
data = aus_plotting.loc[india_trend].loc[:, '1980':'2008'].T
data.plot(kind='line')
plt.show()
```

```python
# Plot Total Immigrants based on DevelopmentRegion

plt.figure(figsize=(10,6))
development = aus_plotting.groupby(by=['DevName'])

developed = development.get_group('Developed regions').Total
developed.plot(kind='line', label = 'Developed Regions')

developing = development.get_group('Developing regions').Total
developing.plot(kind='line', label = 'Developing Regions')

plt.legend()
plt.show()
```

```python
# Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

top_5_trend = aus_plotting.sort_values(by='Total', ascending = False).head()

label = top_5_trend.Country
values = top_5_trend.loc[:, '1980':'2008']

df1 = values.T
df1.columns = label


df1.plot(kind='line')
plt.legend(fontsize = 7)
plt.show()
```

```python
# For the above use area plot with alpha = 0.5


top_5_trend = aus_plotting.sort_values(by='Total', ascending = False).head()

label = top_5_trend.Country
values = top_5_trend.loc[:, '1980':'2008']

df1 = values.T
df1.columns = label


df1.plot(kind='area', stacked = False, alpha = 0.5)
plt.legend(fontsize = 7)
plt.show()
```

```python
# Compare Immigration Trend for top-5 countires using stacked area plot

top_5_trend = aus_plotting.sort_values(by='Total', ascending = False).head()

label = top_5_trend.Country
values = top_5_trend.loc[:, '1980':'2008']

df1 = values.T
df1.columns = label


df1.plot(kind='area', stacked = True, alpha = 0.5)
plt.legend(fontsize = 7)
plt.show()
```
