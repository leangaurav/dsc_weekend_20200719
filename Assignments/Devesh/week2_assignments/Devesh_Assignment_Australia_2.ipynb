- Find total immigrants from each category by area name and plot barchart for each area
- Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
- Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
- Plot Total Immigrants based on DevelopmenRegion 

  
- Plot immigration trend for India using line plot
- Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

- Compare Immigration Trend for top-5 countires using stacked area plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# - Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants

AustraliaData = pd.read_excel('AustraliaFiltered.xlsx')
MaxImmigrants = (AustraliaData.sort_values(by=['Total'], ascending=False)['Total']).head()
MaxImmigrantsC = (AustraliaData.sort_values(by=['Total'], ascending=False)['Country']).head()

plt.figure(figsize=(15,3))
plt.bar(MaxImmigrantsC,MaxImmigrants)
```

```python
#- Find total immigrants from each category by area name and plot barchart for each area
AreaWiseData = AustraliaData.groupby(by = ["AreaName"])
df = AreaWiseData.AreaName.unique() 
areaName = []
areaImmi = []
for i in df:
    areaName.append(i[0])
    total = AustraliaData.loc[AustraliaData['AreaName'] == i[0],'Total'].sum()
    areaImmi.append(total)

areaName = pd.Series(areaName)
areaImmi = pd.Series(areaImmi)

plt.figure(figsize=(14,3))
plt.bar(areaName, areaImmi)
```

```python
#Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
yearWorldDataTemp = AustraliaData.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
yearWorldDataTemp = yearWorldDataTemp.replace(to_replace='..', value=0)
yearWorldDataTemp = yearWorldDataTemp.sum()
yearWorldData = []
for i in yearWorldDataTemp:
    yearWorldData.append(i)

plt.figure(figsize=(14,5))
plt.plot(yearWorldData, label = 'YOY Total')
plt.legend()
plt.show()
```

```python
#Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).

AsiaDataTemp = AustraliaData.loc[AustraliaData['AreaName'] == 'Asia']
AsiaDataTemp = AsiaDataTemp.replace(to_replace='..', value=0)
years = (AsiaDataTemp.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)).sum()
AsiaData = []
for i in years.index:
    AsiaData.append(years[i])
    
AfricaDataTemp = AustraliaData.loc[AustraliaData['AreaName'] == 'Africa']
AfricaDataTemp = AfricaDataTemp.replace(to_replace='..', value=0)
years = (AfricaDataTemp.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)).sum()
AfricaData = []
for i in years.index:
    AfricaData.append(years[i])
    
EuropeDataTemp = AustraliaData.loc[AustraliaData['AreaName'] == 'Europe']
EuropeDataTemp = EuropeDataTemp.replace(to_replace='..', value=0)
years = (EuropeDataTemp.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)).sum()
EuropeData = []
for i in years.index:
    EuropeData.append(years[i])


plt.figure(figsize=(14,5))
plt.plot(AsiaData, label = 'Asia')
plt.plot(AfricaData, label = 'Africa')
plt.plot(EuropeData, label = 'Europe')
plt.legend()
plt.show()
```

```python
#Plot Total Immigrants based on Development Region

developedArea = AustraliaData[AustraliaData['DevName'] == "Developed regions"]['Total']
developingArea = AustraliaData[AustraliaData['DevName'] == "Developing regions"]['Total'] 


plt.figure(figsize=(14,5))
plt.plot(developedArea, label = 'Developed')
plt.plot(developingArea, label = 'Developing')
plt.legend()
plt.show()
```

```python
# Plot immigration trend for India using line plot
IndiaDataTemp = AustraliaData.loc[AustraliaData['Country'] == 'India']
IndiaDataTemp = IndiaDataTemp.replace(to_replace='..', value=0)
years = (IndiaDataTemp.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1))
IndiaData = []
for i in years:
    IndiaData.append(years[i].values)

plt.plot(IndiaData)
```

```python
#Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

MaxImmigrantsC = (AustraliaData.sort_values(by=['Total'], ascending=False)['Country']).head()
MaxImmigrants = (AustraliaData.sort_values(by=['Total'], ascending=False)).head()
MaxImmigrants = MaxImmigrants.replace(to_replace='..', value=0)
MaxImmigrants = MaxImmigrants.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
UK = (MaxImmigrants.values.tolist())[0]
NZ = (MaxImmigrants.values.tolist())[1]
US = (MaxImmigrants.values.tolist())[2]
CHHK = (MaxImmigrants.values.tolist())[3]
CH = (MaxImmigrants.values.tolist())[4]

plt.figure(figsize=(14,5))
plt.plot(UK, label = 'United Kingdom')
plt.plot(NZ, label = 'New Zealand')
plt.plot(US, label = 'United States')
plt.plot(CHHK, label = 'CH Hong Kong')
plt.plot(CH, label = 'China')
plt.legend()
plt.show()
```

```python
#For the above use area plot with alpha = 0.5

MaxImmigrantsC = (AustraliaData.sort_values(by=['Total'], ascending=False)['Country']).head()
MaxImmigrants = (AustraliaData.sort_values(by=['Total'], ascending=False)).head()
MaxImmigrants = MaxImmigrants.replace(to_replace='..', value=0)
MaxImmigrants = MaxImmigrants.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
UK = (MaxImmigrants.values.tolist())[0]
NZ = (MaxImmigrants.values.tolist())[1]
US = (MaxImmigrants.values.tolist())[2]
CHHK = (MaxImmigrants.values.tolist())[3]
CH = (MaxImmigrants.values.tolist())[4]

plt.figure(figsize=(14,5))
plt.plot(UK, label = 'United Kingdom', alpha=0.5)
plt.plot(NZ, label = 'New Zealand', alpha=0.5)
plt.plot(US, label = 'United States', alpha=0.5)
plt.plot(CHHK, label = 'CH Hong Kong', alpha=0.5)
plt.plot(CH, label = 'China', alpha=0.5)
plt.legend()
plt.show()
```

```python
#- Compare Immigration Trend for top-5 countires using stacked area plot

MaxImmigrantsC = (AustraliaData.sort_values(by=['Total'], ascending=False)['Country']).head()
MaxImmigrants = (AustraliaData.sort_values(by=['Total'], ascending=False)).head()
MaxImmigrants = MaxImmigrants.replace(to_replace='..', value=0)
MaxImmigrants = MaxImmigrants.drop(['Unnamed: 0','Country', 'AreaName', 'Region', 'DevName', 'Total'], axis=1)
UK = (MaxImmigrants.values.tolist())[0]
NZ = (MaxImmigrants.values.tolist())[1]
US = (MaxImmigrants.values.tolist())[2]
CHHK = (MaxImmigrants.values.tolist())[3]
CH = (MaxImmigrants.values.tolist())[4]

plt.figure(figsize=(14,5))
plt.fill_between(np.arange(34),UK, alpha=0.5, label = 'United Kingdom')
plt.fill_between(np.arange(34),NZ, alpha=0.5, label = 'New Zealand')
plt.fill_between(np.arange(34),US, alpha=0.5, label = 'United States')
plt.fill_between(np.arange(34),CHHK, alpha=0.5, label = 'CH Hong Kong')
plt.fill_between(np.arange(34),CH, alpha=0.5, label = 'China')
plt.legend()
plt.show()
```
