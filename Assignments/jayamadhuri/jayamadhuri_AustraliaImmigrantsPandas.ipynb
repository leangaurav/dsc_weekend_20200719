```python
# Dataset Analysis
# Describe the dataset
# Use values and index attributes on both Row and Column Indexes
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data

import numpy as np  
import pandas as pd
AustraliaData = pd.read_excel('Australia.xlsx',sheet_name=1,header=20) 
print(AustraliaData.describe())

print()
print(AustraliaData.columns.values)
print()
print(AustraliaData.index.values)

print()
print("Shape",AustraliaData.shape)
print()
print(AustraliaData.dtypes)
```

```python
# Drop
# - Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
# - Find columns for which more than 50% data is missing and drop them (Column Deletion)
# - Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion)
#     Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"

import numpy as np  
import pandas as pd
AustraliaData = pd.read_excel('Australia.xlsx',sheet_name=1,header=20) 
AustraliaData = AustraliaData[AustraliaData['Type'] == "Immigrants"] 
print(AustraliaData['Type'])

print()
AustraliaData = AustraliaData.dropna(thresh=AustraliaData.shape[0]*0.50,how='all',axis=1)
print(list(AustraliaData))

print()
AustraliaData.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], axis = 1, inplace=True) 
print(list(AustraliaData))
```

```python
# Rename
# - Rename columns :OdName -> Country; RegName -> Region
# - Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)
# - Find Total number of immigrants for all years, and create a new column Total in data Frame

AustraliaData = AustraliaData.rename(columns={'OdName':'Country','RegName':'Region'})
print(list(AustraliaData))

print()
AustraliaData.rename(columns=str, inplace=True)
print(list(AustraliaData))

print()
AustraliaData['Total']=AustraliaData.iloc[:,4:39].sum(axis=1)
```
