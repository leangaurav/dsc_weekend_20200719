- Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants
- Find total immigrants from each category by area name and plot barchart for each area
- Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
- Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
- Plot Total Immigrants based on DevelopmenRegion 

  
- Plot immigration trend for India using line plot
- Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
- For the above use area plot with alpha = 0.5


- Compare Immigration Trend for top-5 countires using stacked area plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("AustraliaFiltered.xlsx")
print(df)
```

```python
#Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants

#print(df.sort_values(by='Total',ascending=False).head().Total)
y=df.sort_values(by='Total',ascending=False).head().Total
x=df.sort_values(by='Total',ascending=False).head().Country
#plt.figure(figsize=(15,6))
plt.bar(x,y)
plt.xticks(rotation=90)
plt.show()

tmp=df.sort_values(by='Total',ascending=False).head()
tmp.plot(x='Country',y='Total',kind='bar')
plt.show()
```

```python
#Find total immigrants from each category by area name and plot barchart for each area
grp=df.groupby(by='AreaName')
#print(type(grp.Total.sum()))
grp.Total.sum().plot(kind='bar') #no need to mention x and y as it is a pandas.series
plt.show()

# plt.bar(grp.Total.sum().index,grp.Total.sum().values)
# plt.xticks(rotation=90)
# plt.show()
```

```python
#Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.


#print(df.iloc[:,5:34].sum())
imm=df.iloc[:,5:34].sum()
imm.plot(kind='line')
plt.show()

imm=df.iloc[:,5:34].sum()
plt.plot(imm.index,imm.values)
plt.show()
```

```python
#Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).

##USING DATAFRAME PLOT:
plt.figure(figsize = (15,6))
df1=df.loc[df['AreaName']=='Asia']
trend1=df1.iloc[:,5:34].sum()
#print(list(trend1.index))
trend1.plot(kind='line',label='Asia')
plt.xticks(list(trend1.index),rotation=90)

df2=df.loc[df['AreaName']=='Africa']
trend2=df2.iloc[:,5:34].sum()
trend2.plot(kind='line',label='Africa')


df3=df.loc[df['AreaName']=='Europe']
trend3=df3.iloc[:,5:34].sum()
trend3.plot(kind='line',label='Europe')
plt.legend()
plt.show()

##USING MATPLOTLIB PLOT:
plt.figure(figsize = (15,6))
grp=df.groupby(by='AreaName')
asia=grp.get_group('Asia').iloc[:,5:34].sum()
plt.plot(asia,label='Asia')                   # or asia.plot(kind='line')

africa=grp.get_group('Africa').iloc[:,5:34].sum()
plt.plot(africa,label='Africa')

europe=grp.get_group('Europe').iloc[:,5:34].sum()
plt.plot(europe,label='Europe')
plt.xticks(list(asia.index),rotation=90)
plt.legend()
plt.show()
```

```python
#Plot Total Immigrants based on DevelopmenRegion

plt.figure(figsize=(15,6))
grp=df.groupby(by='DevName')
dev1=grp.get_group('Developing regions').iloc[:,5:34].sum()
dev2=grp.get_group('Developed regions').iloc[:,5:34].sum()
plt.plot(dev1,label='Developing regions')
plt.plot(dev2,label='Developed regions')
plt.legend()
plt.xticks(list(dev1.index),rotation=90)
plt.show()
```

```python
#Plot immigration trend for India using line plot

plt.figure(figsize=(15,6))
grp=df.groupby(by='Country')
trend=grp.get_group('India').iloc[:,5:34].sum()
trend.plot(kind='line',label='India')
plt.legend()
plt.xticks(list(trend.index),rotation=90)
plt.show()
```

```python
#Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

plt.figure(figsize=(15,6))
top=df.sort_values(by='Total',ascending=False).head()
#print(top.iloc[0,5:34])

top.iloc[0,5:34].plot(kind='line',label=top.iloc[0].Country)
plt.plot(top.iloc[1,5:34],label=top.iloc[1].Country)
plt.plot(top.iloc[2,5:34],label=top.iloc[2].Country)
plt.plot(top.iloc[3,5:34],label=top.iloc[3].Country)
plt.plot(top.iloc[4,5:34],label=top.iloc[4].Country)
plt.legend()
plt.show()
```

```python
#For the above use area plot with alpha = 0.5

top.iloc[0,5:34].plot(kind='area',label=top.iloc[0].Country,alpha=0.5)
top.iloc[1,5:34].plot(kind='area',label=top.iloc[1].Country,alpha=0.5)
top.iloc[2,5:34].plot(kind='area',label=top.iloc[2].Country,alpha=0.5)
top.iloc[3,5:34].plot(kind='area',label=top.iloc[3].Country,alpha=0.5)
top.iloc[4,5:34].plot(kind='area',label=top.iloc[4].Country,alpha=0.5)
plt.legend()
plt.show()
```

```python
#Compare Immigration Trend for top-5 countires using stacked area plot

top.iloc[0,5:34].plot(kind='area',label=top.iloc[0].Country,stacked=True)
top.iloc[1,5:34].plot(kind='area',label=top.iloc[1].Country,stacked=True)
top.iloc[2,5:34].plot(kind='area',label=top.iloc[2].Country,stacked=True)
top.iloc[3,5:34].plot(kind='area',label=top.iloc[3].Country,stacked=True)
top.iloc[4,5:34].plot(kind='area',label=top.iloc[4].Country,stacked=True)
plt.legend()
plt.show()
```
